```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ipywidgets as ipy
from ipywidgets import Output

df = pd.read_csv("/Users/lucianoleroi/Desktop/piscine/Jour6/sales.csv",keep_default_na=False)
output3= Output()
with output3:
  
  display(df)

display(output3)
```

```python

camembert=df.groupby("region")["total_price"].sum()
plt.pie(camembert, explode=None, labels=camembert.index,   autopct='%1.1f%%')
plt.title('Répartition des ventes par région')
plt.show()
```

```python
camembert2=df.groupby("category")["total_price"].sum()
plt.pie(camembert2, explode=None, labels=camembert2.index, autopct='%1.1f%%')
plt.title('Répartition des ventes par catégorie')
plt.show()
```

Dans le premier graphique, on observe que la répartition des ventes par région est très inégale. La majorité des ventes sont attribuées à la région EMEA avec 48,4 %, suivie de la région NA avec 38,6 %, et le Japon avec 4,8 % et 8,2% pour APAC. On constate donc que la région EMEA est le plus grand consommateur de voitures.

Par ailleurs, dans le deuxième graphique, on constate que la distribution des ventes entre les catégories est également très inégale. La majorité des achats provient des voitures classiques avec 48 %, suivies par les voitures vintage avec 23 %. Les bus et les camions représentent 11,3 %, les motos 10,7 %, les avions 3,5 %, les trains 1,5 %, et enfin les bateaux 1,2 %. Cette distribution semble tout à fait normale étant donné que le grand public cherche principalement des voitures classiques pour répondre à leurs besoins de déplacement. Les autres catégories répondent à des publics plus spécifiques. Les voitures vintage, bien que plus chères, attirent presque un quart du public en raison de leur valeur esthétique. Les bus et les camions, destinés à un usage commercial, expliquent leur part de 11,3 %. Les motos, répondant à un public plus restreint, représentent 10,7 % des ventes. Enfin, les avions, les bateaux et les trains, bien que minoritaires et ne dépassant pas 6 % cumulés, répondent à un public réduit en raison de leur prix élevé.

Après avoir analysé les graphiques, nous constatons que la répartition du prix total en fonction de la taille des affaires ("dealsize") est proportionnelle. En effet, plus l'affaire est grande, plus le prix total augmente.

Cette tendance est également confirmée par l'analyse des régions : plus la taille de l'affaire est importante, plus le prix total est élevé.

Par ailleurs, les bandes les plus colorées représentent les zones de plus forte concentration de données. Elles illustrent les prix les plus courants pour chaque catégorie. Par exemple, on observe que la majorité des voitures sont vendues entre 700 $ et 2 500 $ (prix total). Ces données révèlent donc les prix les plus fréquents selon les catégories. On remarque également que la répartition des ventes par région se situe entre 1 500 $ et 5 500 $ dans les quatre régions
